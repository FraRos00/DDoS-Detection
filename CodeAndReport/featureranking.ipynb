```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('./cleaned_scaled_dataset.csv',low_memory=False).drop(columns="SimillarHTTP")
features = df.columns[:-1]
labels = df["label"]
df
```

```python
rf_clf = RandomForestClassifier(max_depth=None,criterion='entropy',min_samples_split=3,n_estimators=200,n_jobs=8) 
rf_clf.fit(df[features], labels)
```

```python
importances = rf_clf.feature_importances_
feature_ranking = sorted(zip(features, importances), key=lambda x: x[1], reverse=True)
feature_ranking
```
